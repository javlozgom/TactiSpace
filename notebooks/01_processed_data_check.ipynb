```python
from pathlib import Path
import pandas as pd

# Cargar parquet (ajusta la ruta si lo ejecutas desde otro sitio)
DATA_PATH = Path("data/processed/events.parquet")
if not DATA_PATH.exists():
    # fallback típico si lo ejecutas desde notebooks/
    DATA_PATH = Path("../data/processed/events.parquet")

df = pd.read_parquet(DATA_PATH)

event_counts = df["type"].value_counts(dropna=False)

print("TIPOS DE EVENTO (ordenados por frecuencia):\n")
for event_type, n in event_counts.items():
    print(f"- {event_type}: {n:,}")

print(f"\nTotal tipos distintos: {df['type'].nunique(dropna=True)}")
```
